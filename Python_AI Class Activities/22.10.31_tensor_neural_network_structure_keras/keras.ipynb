```python
from keras.datasets import mnist
```

```python
(train_images, train_labels),(test_images, test_labels) = mnist.load_data()
```

```python
from keras import models
from keras import layers
```

```python
network = models.Sequential() #신경망을 나타내는 네트워크, sequential : 순차적
network.add(layers.Dense(512, activation='relu', input_shape = (28 * 28,))) #네트워크에 Dense 레이어 추가, 모델 입력, 입력되는 데이터 모양을 만들어야한다. ,를 빼면 에러가 난다. 2차원으로 넣어야하기 때문
network.add(layers.Dense(10, activation = 'softmax')) #인풋에서 아웃풋만 만듬
```

```python
network.compile(optimizer='rmsprop', 
                loss='categorical_crossentropy',
                metrics=['accuracy']) #3개의 옵션을 주어야함, optimizer : 성능 최적화(가중치 조정),다중 분류 모델에서는 주로 categorical_crossentropy을 사용
```

```python
# 데이터 타입 변환
train_images = train_images.reshape(60000,28*28)
train_images = train_images.astype('float32') / 255#텐서플로우나 케라스는 실수형으로 바꿔서 해야한다.

test_images = test_images.reshape((10000,28*28))
test_images = test_images.astype('float32')/255
```

```python
# 분류형 데이터의 설정
from tensorflow.keras.utils import to_categorical
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128) #epochs는 반복 횟수, batch_size 처리하는 속도가 빠르지만 메모리가 부족하면 팅김
#loss값 : 오차
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels) # 평가
```

```python
print('test_acc: ', test_acc)
```
