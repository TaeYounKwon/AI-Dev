```python
from azure.storage.file import FileService
file_service = FileService(account_name='*****************',
                           account_key='************************************************')
```

```python
file_service.create_share('myshare') #저장소 만들기
```

```python
file_service.create_directory('myshare','sampledir') #디렉토리 만들기
```

```python
from azure.storage.file import ContentSettings

file_service.create_file_from_path('myshare', None,'myfile', 'MS AI SCHOOL 포스터.jpg', content_settings = ContentSettings(content_type = 'image.jpg'))
```

```python
generator = file_service.list_directories_and_files('myshare')
```

```python
for file_or_dir in generator:
  print(file_or_dir.name)
```

```python
file_service.get_file_to_path('myshare', None, 'myfile','out-MS AI SCHOOL 포스터.jpg')
```
