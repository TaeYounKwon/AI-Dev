```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
print(wine.DESCR)
```

```python
data = wine.data
label = wine.target
columns = wine.feature_names
```

```python
data = pd.DataFrame(data, columns = columns)
data.head()
```

```python
# k-Means
# 데이터 전처리
# 1. MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = scaler.fit_transform(data) 
```

```python
data.shape
```

```python
# PCA(차원의 축소)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
data = pca.fit_transform(data)
```

```python
data
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3) #몇 개의 클러스터로 나눌지 정해야한다
```

```python
kmeans.fit(data)
```

```python
cluster = kmeans.predict(data)
```

```python
cluster
```

```python
plt.scatter(data[:,0],data[:,1], c=cluster, edgecolors='k',  linewidths=1)
```

```python
# Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single') #클러스터의 갯수, 가장 가까운 거리
complete_clustering = AgglomerativeClustering(n_clusters=3, linkage='complete') #가장 먼 거리
average_clustering = AgglomerativeClustering(n_clusters=3, linkage='average') #평균값
```

```python
single_clustering.fit(data)
complete_clustering.fit(data)
average_clustering.fit(data)
```

```python
single_cluster = single_clustering.labels_
complete_cluster = complete_clustering.labels_
average_cluster = average_clustering.labels_
```

```python
print(single_cluster)
print(complete_cluster)
print(average_cluster)
```

```python
plt.scatter(data[:,0],data[:,1],c=single_cluster)
```

```python
plt.scatter(data[:,0],data[:,1],c=complete_cluster)
```

```python
plt.scatter(data[:,0],data[:,1],c=average_cluster)
```

```python
plt.scatter(data[:,0],data[:,1],c=label) #정답지, 비교해본 결과 average_cluster가 가장 정확
```

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(10,10))
children = single_clustering.children_
distance = np.arange(children.shape[0])
no_of_observations = np.arange(2, children.shape[0]+2)
linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)
dendrogram(linkage_matrix, p=len(data), labels = single_cluster, show_contracted = True, no_labels = True)
```

```python
# Silhouette 비지도학습 평가
from sklearn.metrics import silhouette_score

best_n = 1
best_score = -1
for n_cluster in range(2,11):
  kmeans = KMeans(n_clusters=n_cluster)
  kmeans.fit(data)
  cluster = kmeans.predict(data)
  score = silhouette_score(data, cluster)
  print('클러스터의 수 : {} 실루엣 점수 : {:.2f}'.format(n_cluster, score))

  if score > best_score:
    best_n = n_cluster
    best_score = score

print('가장 높은 실루엣 점수를 가진 클러스터 수: {}, 실루엣 점수{:.2f}'.format(best_n,best_score))
```

```python
# Silhouette 비지도학습 평가
from sklearn.metrics import silhouette_score

best_n = 1
best_score = -1
for n_cluster in range(2,11):
  average_clustering = AgglomerativeClustering(n_clusters=n_cluster,linkage = 'average')
  average_clustering.fit(data)
  cluster = average_clustering.labels_ #average_clustering은 label로 바로작성
  score = silhouette_score(data, cluster)
  print('클러스터의 수 : {} 실루엣 점수 : {:.2f}'.format(n_cluster, score))

  if score > best_score:
    best_n = n_cluster
    best_score = score

print('가장 높은 실루엣 점수를 가진 클러스터 수: {}, 실루엣 점수{:.2f}'.format(best_n,best_score))
```
