```python
#보스턴 집값 데이터셋
from keras.datasets import boston_housing
(train_data, train_labels),(test_data, test_labels) = boston_housing.load_data()
```

```python
from keras import models
from keras import layers

#신경망을 함수로 만들기
def build_model():
  model = models.Sequential()
  model.add(layers.Dense(64, activation='relu',input_shape = (train_data.shape[1],))) #(train_data.shape[1]):학습용 데이터의 갯수
  model.add(layers.Dense(64, activation='relu'))
  model.add(layers.Dense(1))#집값의 데이터만 구하면 되기에 1개 activation의 디폴트는 identity function

  model.compile(optimizer = 'rmsprop',
                loss = 'mse',
                metrics = ['mae'])
  return model
```

```python
#데이터가 별로 없어서 K-folder검증을 거쳐야한다.
#K-folder 검증
k = 4 #4개로 나눈다
num_val_samples = len(train_data) // k #폴더의 사이즈
all_scores = []

for i in range(k):
  print('처리중인 폴드 #',i)

  # 검증 데이터 준비 : k번째 분할
  val_data = train_data[i * num_val_samples :(i+1)*num_val_samples]
  val_labels = train_labels[i*num_val_samples :(i+1)*num_val_samples]

  # 훈련 데이터 준비
  partial_train_data = np.concatenate([   #concatenate() : 리스트를 합쳐줌
    train_data[: i * num_val_samples],
    train_data[(i+1)*num_val_samples :]], axis=0)
  
  partial_train_labels = np.concatenate([   #concatenate() : 리스트를 합쳐줌
    train_labels[: i * num_val_samples],
    train_labels[(i+1)*num_val_samples :]], axis=0)

  model = build_model()
  model.fit(partial_train_data, partial_train_labels, epochs = 20, batch_size=1, verbose=0) #폴더가 4개이기에 80번 돌아간다
  
  val_mse, val_mae = model.evaluate(val_data, val_labels,verbose=0)
  all_scores.append(val_mae)
```

```python
#K-folder 검증
k = 4 #4개로 나눈다
num_epochs = 500

num_val_samples = len(train_data) // k #폴더의 사이즈
all_scores = []

for i in range(k):
  print('처리중인 폴드 #',i)

  # 검증 데이터 준비 : k번째 분할
  val_data = train_data[i * num_val_samples :(i+1)*num_val_samples]
  val_labels = train_labels[i*num_val_samples :(i+1)*num_val_samples]

  # 훈련 데이터 준비
  partial_train_data = np.concatenate([   #concatenate() : 리스트를 합쳐줌
    train_data[: i * num_val_samples],
    train_data[(i+1)*num_val_samples :]], axis=0)
  
  partial_train_labels = np.concatenate([   #concatenate() : 리스트를 합쳐줌
    train_labels[: i * num_val_samples],
    train_labels[(i+1)*num_val_samples :]], axis=0)

  model = build_model()
  history = model.fit(partial_train_data, partial_train_labels, epochs = num_epochs, batch_size=1,validation_data = (val_data, val_labels), verbose=0)

  mae_history = history.history['mae']
  
  all_scores.append(mae_history)
```

```python
average_mae_history = [np.mean([x[i] for x in all_scores]) for i in range(num_epochs)]
```

```python
plt.plot(range(1, len(average_mae_history)+1), average_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
```
