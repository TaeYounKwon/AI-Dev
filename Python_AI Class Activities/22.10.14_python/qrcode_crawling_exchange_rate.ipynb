```python
import qrcode
with open('site_list.txt','r', encoding='UTF8') as f: #with open:블록을 생성하고, 벗어나면 파일을 닫아줌
  read_lines = f.readlines() #readlines() 전체 내용을 한번에 읽음
  for line in read_lines:
    line = line.strip()#strip() 글자 이외를 날려줌
    print(line)

    qr_data = line
    qr_image = qrcode.make(qr_data)
    qr_image.save(qr_data + '.png')
```
