```python
# 선형적으로 구분되지 않은 데이터 차원 축소
# Import all the necessary Library
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_circles
from sklearn.decomposition import TruncatedSVD
```

```python
# 선형적으로 구분되지 않느 데이터 생성
# Create the dataset that are not a linear model
# By using make_circles, can create the datasets that are like circle coordinate
features, _ = make_circles(n_samples=1000, random_state=1, noise=0.1, factor=0.1 )
print(features)
print("Number of Original Feature: ", features.shape[1])
print("Len of features: ", len(features))
```

```python
# 방사 기저 함수(radius basis function, RBF)를 사용하여 커널에 PCA 적용
kpca = KernelPCA(kernel="rbf", gamma=15, n_components=1)
features_kpca = kpca.fit_transform(features)
```

참조! => KernelPCA 여러 방법들이 많이 있음, 그중
rbf = 가우시안 방사

```python
print("Number of Original Feature: ", features.shape[1])
print("Number of Reduced Feature: ", features_kpca.shape[1])
print("Len of Reduced Feature: ", len(features_kpca))
```

```python
features, _ = make_circles(n_samples=1000, random_state=1, noise=0.1, factor=0.1 )
print(features)
print("Number of Original Feature: ", features.shape[1])
print("Len of features: ", len(features))
```

```python
# SVD를 사용하여 커널에 PCA 적용
tsvd = TruncatedSVD(n_components=1,n_iter=10,random_state=42)
features_tsvd = tsvd.fit_transform(features)
```

```python
print("Number of Original Feature: ", features.shape[1])
print("Number of Reduced Feature: ", features_tsvd.shape[1])
print("Len of Reduced Feature: ", len(features_tsvd))
```
