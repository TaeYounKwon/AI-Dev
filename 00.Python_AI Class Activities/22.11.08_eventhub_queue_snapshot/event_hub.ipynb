```python
from azure.eventhub.aio import EventHubProducerClient #데이터를 이벤트허브로 쏠 수 있게함
from azure.eventhub import EventData#데이터를 이벤트 허브로 쏜다.
```

```python
eventhub = EventHubProducerClient.from_connection_string(conn_str= '****************************************************************************')
#EventHubProducerClient을 연결문자열을 이용하여 포탈과 연결, 빈칸에 키값 넣기
```

```python
import time #시간을 표시하기 위해
from datetime import datetime#현재 시간을 보낸다

i = 0 #반복 수 재는 변수
while(1):
  message = 'time: '+ str(datetime.now()) # 현재 시간, str으로 변환
  eventhub.send_event(EventData(message)) # 전송하는 코드, send_event는 하나만 보냄, 여러개 보낼때 send_batch
  
  i+=1
  if i%10==0:  #10번에 한번씩 보냄 즉 20초마다 보내게 된다.
    print('message {} sending...'.format(i))

  time.sleep(2) #2초마다 보냄
```
