```python
import os, uuid # uuid:임시변수 패키지
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
```

```python
# Blob 서비스에 연결
# 접속을 위한 주소, 키값 불러오기
#azuer active key 사용 권장, 일반 키값으로 가져오면 키 유출됨, 연결 문자열 키 복사
connect_string = '***********************************************************************************************************************'
# 블랍 서비스-컨테이너-블랍 순으로 접근
# 블랍 서비스 연결, 블랍서비스에 연결한 컨테이너 서비스 생성
blob_service_client = BlobServiceClient.from_connection_string(connect_string)
```

```python
# Container 생성
#컨테이너이름을 겹치지 않는 난수로 만들기위해 uuid사용
container_name = str(uuid.uuid4())
container_client = blob_service_client.create_container(container_name)
```

```python
# 데이터 업로드
local_path = './data'
os.mkdir(local_path)
local_file_name = str(uuid.uuid4()) + '.txt'
upload_file_path = os.path.join(local_path, local_file_name) #로컬 파일 경로 세팅
```

```python
# 업로드 할 파일 준비
file = open(file=upload_file_path, mode = 'w')
file.write('Hello Azure Storage')
file.close()

blob_client = blob_service_client.get_blob_client(  #새로운 블랍 클라이언트를 만들기
    container = container_name, #어느 컨테이너에 파일을 올릴것인가
    blob = local_file_name #업로드 할 파일의 정보
) 
# 파일 업로드
with open(file = upload_file_path, mode = 'rb') as data: # rb : 바이너리 형태로 읽기, 텍스트로 읽으면 이미지나 다른 파일은 깨짐
  blob_client.upload_blob(data) #데이터를 가지고 업로드 블랍
```

```python
# 파일 목록의 확인
blob_list = container_client.list_blobs()
for blob in blob_list:
  print('\t' + blob.name)
```

```python
# 업로드된 파일의 다운로드
download_file_path = os.path.join(local_path,
                                 str.replace(local_file_name,'.txt','DOWNLOAD.txt'))
container_client = blob_service_client.get_container_client(container = container_name)
with open(file=download_file_path, mode='wb') as download_file:
  download_file.write(container_client.download_blob(blob.name).readall()) # 파일을 쓰는 작업
```

```python
# 실습한 자원 정리
print('Press the Enter key to beigin clean up')
input()

print('Deleting blob container...')
container_client.delete_container()

print('Deleting the local source and downloaded files... ')
os.remove(upload_file_path)
os.remove(download_file_path)
os.rmdir(local_path) #디렉토리 삭제
print('Done')
```
