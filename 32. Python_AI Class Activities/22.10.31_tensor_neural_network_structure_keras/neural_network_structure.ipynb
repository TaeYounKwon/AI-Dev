```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# AND 게이트
def AND(a, b):
  input = np.array([a,b])

  #가중치 설정
  weights = np.array([0.4,0.4])
  bias = -0.6

  #출력값
  value = np.sum(input * weights) + bias

  #반환값
  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
# OR 게이트
def OR(a, b):
  input = np.array([a,b])

  #가중치 설정
  weights = np.array([0.4,0.4])
  bias = -0.3

  #출력값
  value = np.sum(input * weights) + bias

  #반환값
  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
# NAND 게이트
def NAND(a, b):
  input = np.array([a,b])

  #가중치 설정
  weights = np.array([-0.6,-0.6])
  bias = 0.7

  #출력값
  value = np.sum(input * weights) + bias

  #반환값
  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(x1, x2)

  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

```python
#Activation Function
#Step Function(계단 함수)
def step_function(x):
  if x>0:
    return 1
  else:
    return 0
```

```python
#step function 조건문 없이
def step_function_for_numpy(x):
  y = x > 0
  value = y.astype(np.int) #boolean값을 int형으로 변환
  
  return value
```

```python
print(step_function(-3))
print(step_function(5))
```

```python
# 시그모이드 함수

#np.exp() : 지수함수, 숫자가 한번에 커지기 때문에 1~10사이만 넣기
def sigmoid(x):
  value = 1 / (1+np.exp(-x)) #시그모이드 함수
  return value

print(sigmoid(3))
print(sigmoid(-3))
```

```python
plt.grid()
x = np.arange(-5,5,0.01)
y1 = sigmoid(x)
y2 = step_function_for_numpy(x)

plt.plot(x,y1,'r-')
plt.plot(x,y2,'b--')
```

```python
# Softmax(a) 원리 파악하기 위한 예시
def Softmax(a):
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a/ sum_exp_a

  return y
```

```python
a = np.array([0.3, 0.2, 3.0, -1.2])
print(Softmax(a))
print(np.sum(Softmax(a))) #전체의 합은 1이다.
```
