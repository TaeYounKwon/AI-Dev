```python
#file share와 연결
from azure.storage.file import FileService
file_service = FileService(account_name='***************',
                           account_key='************************************************')
```

```python
file_service.create_share('myshare') #저장소 만들기
```

```python
file_service.create_directory('myshare','sampledir') #sampledir 디렉토리 만들기
```

```python
#capybara.png를 공유파일에 복사
from azure.storage.file import ContentSettings

file_service.create_file_from_path('myshare', None,'myfile', 'capybara.png', content_settings = ContentSettings(content_type = 'image.png'))
```

```python
generator = file_service.list_directories_and_files('myshare')
for file_or_dir in generator:
  print(file_or_dir.name)
```

```python
#스냅샷 찍기
metadata = {'foo':'bar'}
snapshot = file_service.snapshot_share('myshare', metadata=metadata)
```

```python
#전체스냅샷의 목록을 가져온다
shares = list(file_service.list_shares(include_snapshots=True))
for s in shares:
  print(s)
```

```python
#파일과 디렉토리 목록 출력
directories_and_files = list(
    file_service.list_directories_and_files('myshare',snapshot='2022-11-08T01:09:00.0000000Z')
)
```

```python
#스냅샷 지우기
file_service.delete_share('myshare', snapshot='2022-11-08T01:09:00.0000000Z')
```

```python
#공유폴더내의 파일 삭제, 
file_service.delete_file('myshare',None, 'myfile')
```

```python
#공유 폴더 삭제
file_service.delete_share('myshare')
```
